```python
import matplotlib
%matplotlib inline
```

```python
from utils.plot import plot_reward

env_ids = [env + 'NoFrameskip-v4' for env in ['AirRaid', 'Assault', 'Carnival', 'DemonAttack']]

for env_id in env_ids:
    log_dir = 'log/' + env_id + '/'
    plot_reward(log_dir, env_id, 'DQN', 1000000, bin_size=10, smooth=1, time=0, ipynb=True)
```

```python
import gym

for env_id in env_ids:
    env = gym.make(env_id)
    print('{:25s}'.format(env_id), env.action_space, env.env.get_action_meanings())
```
